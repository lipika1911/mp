```python
import time
import os  # Import the os module
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Measure start time for setup
setup_start_time = time.time()
print(f"Setup start time: {time.ctime(setup_start_time)}")

# Define dataset paths
data_dir = "./Fruits-360/Training/"

# List categories
categories = os.listdir(data_dir)

# Set image size
img_size = (100, 100)

# Build the CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_size[0], img_size[1], 3)),
    MaxPooling2D(pool_size=(2, 2)),
    
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(categories), activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Display the model architecture
model.summary()

# Measure end time for setup
setup_end_time = time.time()
print(f"Setup end time: {time.ctime(setup_end_time)}")
setup_elapsed_time = setup_end_time - setup_start_time
print(f"Total setup time: {setup_elapsed_time:.2f} seconds")
```

```python
# Measure start time for data generation
data_gen_start_time = time.time()
print(f"Data generation start time: {time.ctime(data_gen_start_time)}")

# Set up data generators
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(100, 100),
    batch_size=16,  # Reduced batch size
    class_mode='categorical',
    subset='training'
)

validation_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(100, 100),
    batch_size=16,  # Reduced batch size
    class_mode='categorical',
    subset='validation'
)

# Measure end time for data generation
data_gen_end_time = time.time()
print(f"Data generation end time: {time.ctime(data_gen_end_time)}")
data_gen_elapsed_time = data_gen_end_time - data_gen_start_time
print(f"Total data generation time: {data_gen_elapsed_time:.2f} seconds")
```

```python
# Measure start time for training
train_start_time = time.time()
print(f"Training start time: {time.ctime(train_start_time)}")

# Train the model
history = model.fit(train_generator, validation_data=validation_generator, epochs=15)

# Measure end time for training
train_end_time = time.time()
print(f"Training end time: {time.ctime(train_end_time)}")

# Calculate and print training time
train_elapsed_time = train_end_time - train_start_time
print(f"Total training time: {train_elapsed_time:.2f} seconds")
```

```python
# Measure start time for visualization
plot_start_time = time.time()
print(f"Visualization start time: {time.ctime(plot_start_time)}")

# Visualize training results
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.show()

# Measure end time for visualization
plot_end_time = time.time()
print(f"Visualization end time: {time.ctime(plot_end_time)}")

# Calculate and print visualization time
plot_elapsed_time = plot_end_time - plot_start_time
print(f"Total visualization time: {plot_elapsed_time:.2f} seconds")
```

```python
# Measure start time for evaluation
eval_start_time = time.time()
print(f"Evaluation start time: {time.ctime(eval_start_time)}")

# Evaluate the model on test data
test_loss, test_acc = model.evaluate(validation_generator, verbose=2)
print(f"Test Accuracy: {test_acc:.4f}")

# Measure end time for evaluation
eval_end_time = time.time()
print(f"Evaluation end time: {time.ctime(eval_end_time)}")

# Calculate and print evaluation time
eval_elapsed_time = eval_end_time - eval_start_time
print(f"Total evaluation time: {eval_elapsed_time:.2f} seconds")
```

```python
import numpy as np
import cv2
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt

# Load the trained model
# model = load_model('path_to_your_trained_model.h5')

def preprocess_image(img_path, target_size):
    img = image.load_img(img_path, target_size=target_size)
    img_array = image.img_to_array(img)
    img_array = img_array / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    return img_array

# Path to your test image
img_path = './test_img7.jpg'
target_size = (100, 100)  # Modify as needed based on your model input

# Preprocess the image
preprocessed_image = preprocess_image(img_path, target_size)

# Make a prediction
prediction = model.predict(preprocessed_image)
predicted_class_index = np.argmax(prediction, axis=1)
confidence = np.max(prediction)  # Confidence of the prediction

# Get class labels (ensure this matches your training labels)
class_labels = list(test_generator.class_indices.keys())

# Get the predicted class label
predicted_class_label = class_labels[predicted_class_index[0]]

# Load and display the image
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB for displaying with matplotlib

plt.imshow(img)
plt.title(f"Predicted Class: {predicted_class_label}")
plt.axis('off')
plt.show()

# Print results
print(f"Predicted Class: {predicted_class_label}")
print(f"Prediction Accuracy: {confidence:.4f}")
```
